```r
# Computando grupos 
X = c(3, 4, 5, 6, 7)
Y = c(1, 3, 5, 7, 9)
Z = c(5, 5, 5, 5, 5)
W = c(3, 5, 5, 5, 7)
V = c(3, 5, 5, 6, 6)
```

```r
# Calculando o menor e maior valor dos grupos
Rx = range(X)
Ry = range(Y)
Rz = range(Z)
Rw = range(W)
Rv = range(V)
```

```r
# Calculando a amplitude dos grupos
Ax = diff(Rx)
Ay = diff(Ry)
Az = diff(Rz)
Aw = diff(Rw)
Av = diff(Rv)
```

```r
# Printando as amplitudes
print(Ax)
print(Ay)
print(Az)
print(Aw)
print(Av)
```

```r
# Desvio padrão 
pD = c(45, 74, 56, 98, 32, 97)
sD = c(48, 81, 33, 40, 84)
```

```r
# Função para o cálculo do coeficiente de variação
cv = function(variable){sd(variable)/mean(variable)}
```

```r
# Contagem de peixes
peixes = c(76, 102, 12, 39, 55, 93, 98, 53, 102)
cv(peixes)*100
```
