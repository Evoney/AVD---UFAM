```r
# Teste de hip√≥tese

x <- c(6.2, 6.6, 7.1, 7.4, 7.6, 7.9, 8.0, 8.3, 8.4, 8.5, 8.6, 8.8,
8.8, 9.1, 9.2, 9.4, 9.4, 9.7, 9.9, 10.2, 10.4, 10.8, 11.3, 11.9)
```

```r
# teste de normalidade para alpha = 0.05
shapiro.test(x)
```

```r
# t teste
t.test(x, mu=9, alternative="two.sided", conf.level=0.95)
```
