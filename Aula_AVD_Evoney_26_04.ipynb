```r
x = c(16, 17, 18, 19, 20, 21, 22)
p_x = c(0.05, 0.10, 0.15, 0.25, 0.20, 0.15, 0.10)
```

```r
# Encontrando a média teórica

E_x = 0

for (i in 1:length(x)){
    E_x = E_x + x[i]*p_x[i]
}

print(E_x)
```

```r
# Encontrando a variância

E_x2 = 0

for (i in 1:length(x)){
    E_x2 = E_x2 + x[i]*x[i]*p_x[i]
}


var_x = E_x2 - E_x*E_x

print(var_x)
```

```r
# Para resolução desse exercício será utilizada a função dgeom(x, prob, ...) para variáveis com distruição geométrica
# onde (x) corresponde ao número de falhas ocorridas em uma sequência de Bernoulli antes do sucesso, 
# e (prob) a probabilidade de sucesso em cada tentativa.
```

```r
# para 1 vez ou seja x = 0
dgeom(0, 0.9)
```

```r
# para 2 vezes ou seja x = 1
dgeom(1, 0.9)
```

```r
# para 3 vezes ou seja x = 2
dgeom(2, 0.9)
```
